## 1. Split to train and test

```python
from sklearn.model_selection import train_test_split
X = data.iloc[:, 1:112]
y = [data["ZSN"], data["FIBR_PREDS"], data["P_IM_STEN"], data["REC_IM"], data["OTEK_LANC"]]

X_train, X_test, y_train, y_test = [], [], [], []

# ! Caution
# i = 0 => ZSN or Chronic heart failure
# i = 1 => FIBR_PREDS or Atrial fibrillation
# i = 2 => P_IM_STEN or Post-infarction angina
# i = 3 => REC_IM or Relapse of the myocardial infarction
# i = 4 => OTEK_LANC or Pulmonary edema

for i in range(len(y)):
    # Separate train/test split for each target variable
    X_train_i, X_test_i, y_train_i, y_test_i = train_test_split(X, y[i], train_size=0.8, random_state=0, shuffle = False)
    X_train.append(X_train_i)
    X_test.append(X_test_i)
    y_train.append(y_train_i)
    y_test.append(y_test_i)
```

### Delete outliers

```python
#numbers of values of X_train[0] before removing outliers
for continuous_feature in continuous_features:
  print(f"{continuous_feature} values: {len(X_train[0][continuous_feature])}")
```

```python
#delete all outliers in X_train[0] for the continuous features and Y_train[0] and print numbers of values of X_train[0] and Y_train[0] after removing outliers
from scipy import stats

# Define continuous features
continuous_features = ["AGE", "S_AD_ORIT", "D_AD_ORIT", "K_BLOOD", "NA_BLOOD", "ALT_BLOOD", "AST_BLOOD", "L_BLOOD", "ROE"]

# Calculate IQR for each continuous feature in X_train[0]
Q1 = X_train[0][continuous_features].quantile(0.25)
Q3 = X_train[0][continuous_features].quantile(0.75)
IQR = Q3 - Q1

# Only keep rows in X_train[0] that do not have outliers
X_train_no_outliers = X_train[0][~((X_train[0][continuous_features] < (Q1 - 1.5 * IQR)) | (X_train[0][continuous_features] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Get the indices of X_train_no_outliers
index_no_outliers = X_train_no_outliers.index

# Only keep rows in y_train[0] that do not have outliers
y_train_no_outliers = y_train[0][y_train[0].index.isin(index_no_outliers)]

# Print the number of rows in X_train_no_outliers and y_train_no_outliers
print("Number of rows in X_train[0] after removing outliers: ", X_train_no_outliers.shape[0])
print("Number of rows in y_train[0] after removing outliers: ", y_train_no_outliers.shape[0])
```

### Filling missing values

```python
def fill_missing_values(data):
  _data = data.drop(columns=["IBS_NASL", "S_AD_KBRIG", "D_AD_KBRIG", "KFK_BLOOD", "NA_KB", "NOT_NA_KB", "LID_KB"])
  for categorical_feature in categorical_features:
    _data[categorical_feature].fillna(_data[categorical_feature].median(), inplace = True)
  for continuous_feature in continuous_features:
    _data[continuous_feature].fillna(_data[continuous_feature].mean(), inplace = True)
  return _data
```

```python
X_train[0] = fill_missing_values(X_train[0])
X_train[1] = fill_missing_values(X_train[1])
X_train[2] = fill_missing_values(X_train[2])
X_train[3] = fill_missing_values(X_train[3])
X_train[4] = fill_missing_values(X_train[4])
```

##### Elbow method

```python
# Scree plot
pca_before = decomposition.PCA()
pca_before.fit(X_train_select)
explained_variance_ratio = pca_before.explained_variance_ratio_

plt.plot(explained_variance_ratio)
plt.plot(9, 0.017, marker="o", color="red") # x = 9 => elbow
plt.text(9, 0.019, "A(9, y)")
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.show()
```

##### Kaiser criterion

```python
pca_before = decomposition.PCA()
pca_before.fit(X_train_select)
eigenvalues = pca_before.explained_variance_
num_components_kaiser = np.count_nonzero(eigenvalues > 1)
print(f"Number of components by Kaiser Criterion (> 1 eigenvalue): {num_components_kaiser}")
```

```python
n_c = 44
pca_after = decomposition.PCA(n_components= n_c)
X_pca_kaiser = pca_after.fit(X_train_select)
fig, ax = plt.subplots(figsize = (6, 5))
plt.xlabel('Number of components')
plt.ylabel('Percentage of Variance')
plt.title('Explained variance with PCA')
ax.plot(np.linspace(1, n_c, n_c), np.cumsum(pca_after.explained_variance_ratio_))
ax.plot(np.linspace(1, n_c, n_c), pca_after.explained_variance_ratio_, marker = '.', linestyle = '', color = 'r')
ax.plot([0.8, 3], [np.cumsum(pca_after.explained_variance_ratio_)[2], np.cumsum(pca_after.explained_variance_ratio_)[2]], c = 'c', linestyle = '--')
print(int(np.cumsum(pca_after.explained_variance_ratio_)[2]*100))
ax.legend(('Cumulative explained variance ratio', 'Explained variance ratio', 'Threshold (0.9)'), loc = 7)
ax.legend(('Cumulative explained variance ratio', 'Explained variance ratio'), loc = 7)
ax.set_ylim(0,1)
ax.set_xlim(0.8, n_c + 0.2)
plt.show()
```

##### Conclusion
So both solutions lead to the number of components is 9

```python
pca_after = decomposition.PCA(n_components=9)
X_pca = pca_after.fit_transform(X_train_select)
pca_component_names = [f"PC{i+1}" for i in range(pca_after.n_components_)]
X_train_pca = pd.DataFrame(X_pca, columns = pca_component_names)
X_train_pca
```
